Proveď následující úkoly a zodpověz předložené otázky.

a. Načti dva datové sety studentů do oddělených pandas DataFrame a pomocí funkce concat je spoj do jednoho setu.

```python
import pandas

# Otevření stažených csv
studenti1 = pandas.read_csv("studenti1.csv")
studenti2 = pandas.read_csv("studenti2.csv")

# Alternativa: Načtení csv přímo z kodim.cz
# studenti1 = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/studenti1.csv")
# studenti2 = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/studenti2.csv")

print(len(studenti1))
print(len(studenti2))

studenti = pandas.concat([studenti1, studenti2], ignore_index=True)
print(len(studenti))
studenti
```

b. Pokud studentovi chybí ročník, znamená to, že již nestuduje.

Pokud mu chybí číslo skupiny, znamená to, že jde o dálkového studenta. 

Kolik studentů v datovém setu již nestuduje a kolik jsou dálkoví studenti?

```python
print( studenti["ročník"].isnull().sum() )  # Lze i jako sum(studenti["ročník"].isnull())
print( studenti["kruh"].isnull().sum()   )

# Delší alternativa přes výběr nestudentů a dálkařů:
# len(studenti[studenti["ročník"].isnull()])
# len(studenti[studenti["kruh"].isnull()])
```

c. Vyčisti data od studentů, kteří nestudují nebo studují jen dálkově. Nadále budeme pracovat pouze s prezenčními studenty.

```python
# Při předpokladu, že NaN se nevyskytuje jinde než ve sloupcích "ročník" a "kruh":
# dropna() ve výchozím nastavení odebere řádky, které obsahují aspoň jeden NaN
# studenti = studenti.dropna()
# studenti

# ... Jenže pokud by se NaN vyskytoval i jinde, mohli bychom nechtěně přijít i o další řádky
# Takže "Bezpečnější" alternativa založená jen na vybraných sloupcích:
nestudenti = studenti["ročník"].isnull()
dalkari = studenti["kruh"].isnull()
studenti = studenti[~nestudenti & ~dalkari]
studenti

# Případně to stejné, ale s opačnou logikou pomocí notnull():
# studujici = studenti["ročník"].notnull()
# prezencni = studenti["kruh"].notnull()
# studenti = studenti[studujici & prezencni]
# studenti
```

d. Zjisti, kolik prezenčních studentů je v každém z oborů.

```python
# studenti.groupby("obor").count()
studenti.groupby("obor")["jméno"].count()
# studenti.groupby("obor").size()

# Pro doplnění: count() nezapočítává záznamy s NaN, size() ano
```

e. Zjisti průměrný prospěch studentů v každém oboru.

```python
studenti.groupby("obor")["prospěch"].mean()
# Lze aplikovat i další metody, např. sum, max, min, median, std, first, last, ...
```

```python
# Případně i více statistik najednou
studenti.groupby("obor").agg({"prospěch": ["count", "mean", "median", "min", "max"]})
```

```python
# Nebo vybrané výchozí statistiky metodou describe()
studenti.groupby("obor")["prospěch"].describe()
```

f. Načti datový set s křestními jmény. 

Proveď join s tabulkou studentů tak, abychom věděli pohlaví jednotlivých studentů.

```python
jmena = pandas.read_csv("jmena.csv")
# jmena = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/jmena.csv")

print(jmena.columns)
print(studenti.columns)

studenti_jmena = pandas.merge(studenti, jmena)
studenti_jmena
```
